```python
import pandas as pd
```

```python
men=pd.read_csv("ebay_mens_perfume.csv")
women=pd.read_csv("ebay_womens_perfume.csv")
```

```python
#adding a category column to each dataset(to differentiate gender)
men["Category"] = "Men"
women["Category"]="Women"
```

```python
#Combining the Dataset
sales=pd.concat([men,women], ignore_index=True)
sales
```

```python
#checking for missing value
sales.isnull().sum()
```

```python
#filling nan with 0
sales.fillna(0,inplace=True)
```

```python
#removing duplicates
sales.drop_duplicates()
```

### Changing the available column to integer

```python
sales["available"]=sales["available"].astype("int")
```

### Changing the Columns to Upper Case

```python
sales.columns=sales.columns.str.upper()
```

```python
sales
```

```python
sales.to_csv("Perfume Sales.csv", index=False)
```
